Implement the Batch Gradient Descent (BGD) procedure

```python
import numpy as np
import random
from sklearn.datasets import make_regression 
import pylab
from scipy import stats



def gradient(learning_rate, x, y, epochs,theta):
    loss_list12 = []
    m = x.shape[0] # number of samples
    x_transpose = x.transpose()
    for iter in range(0, epochs):
        hypothesis = np.dot(x, theta)
        loss = hypothesis - y
        J = np.sum(loss ** 2) / (2 * m)  # cost
        print("iter %s | J: %.3f" % (iter, J))
        grad = np.dot(x_transpose, loss) / m         
        theta = theta - learning_rate * grad
        loss_list12.append(J)
    return theta, J,iter,grad,loss_list12
       

def Batch_Gradient_Procedure(x,y,learning_rate,epochs,gradient,theta):
   m, n = np.shape(x)
   x = np.c_[ np.ones(m), x] # insert column
   thetaq,J,iter,grad,losslist12 = gradient(learning_rate, x, y, epochs,theta)
   return thetaq, J,iter,grad,losslist12

x, y = make_regression(n_samples=100, n_features=1, n_informative=1, random_state=0, noise=35) 
learning_rate = 0.01 # learning rate
epochs = 1000
theta = np.ones(2)
x.shape = x.shape
y.shape = y.shape
theta1, loss,iter,grad,losslist12 = Batch_Gradient_Procedure(x,y,learning_rate,epochs,gradient,theta)

print('initial_theta',theta1[0])
print('final_theta:', theta1[1])
print('final_iter',iter,'final_loss:',loss)


```

4. Implement the Stochastic Gradient Descent (SGD) procedure. Redo steps B-F in part 3 and comment on how this procedure may yield different results compared to BGD.

```python
def loss_function(theta,X,y):
 m = len(y)
 predictions = X.dot(theta)
 loss = (1/2*m) * np.sum(np.square(predictions-y))
 return loss
```

```python
def stocashtic_gradient_descent(x, y, theta, learning_rate, epochs):
    m = len(y)
    cost_history = np.zeros(epochs)
    for it in range(epochs):
        cost = 0
        for i in range(m):
            rand_ind = np.random.randint(0, m)
            X_i = x[rand_ind, :].reshape(1, x.shape[1])
            y_i = y[rand_ind].reshape(1, 1)
            prediction = np.dot(X_i, theta)
            theta = theta - (1 / m) * learning_rate * (X_i.T.dot((prediction - y_i)))
            cost += loss_function(theta, X_i, y_i)
        cost_history[it] = cost
    return theta, cost_history
```

5. Implement the Mini-batch Gradient Descent (MBGD) procedure where the batch size is an input. Redo steps B-F in part 3 for two different batch sizes, and comment on how this procedure may yield different results compared to SGD and BGD.


```python
def minibatch_gradient_descent(X,y,theta,learning_rate=0.01, iterations=10, batch_size =20):
    m = len(y)
    cost_history = np.zeros(iterations)
    n_batches = int(m/batch_size)
    for it in range(iterations):
        cost =0.0
        indices = np.random.permutation(m)
        X = X[indices]
        y = y[indices]
        for i in range(0,m,batch_size):
            X_i = X[i:i+batch_size]
            y_i = y[i:i+batch_size]
            X_i = np.c_[np.ones(len(X_i)),X_i]
            prediction = np.dot(X_i,theta)
            theta = theta -(1/m)*learning_rate*( X_i.T.dot((prediction - y_i)))
            cost += loss_function(theta,X_i,y_i)
        cost_history[it]  = cost
    return theta, cost_history
```
